```python
"""
FLC: Speed control of a vehicle
Let two fuzzy inputs (speed difference (SD) and acceleration (A)) and one fuzzy
output throttle control (TC) be there.
X: Universe of discourse [0,240]
Partitions: 
    NL: Open left MF (a = 30, b = 60) 
    NM: Traingular(a = 30, b = 60, c = 90)
    NS: Traingular(a = 60, b = 90, c = 120)
    ZE: Traingular(a = 90, b = 120, c = 150)
    PS: Traingular(a = 120, b = 150, c = 180)
    PM: Traingular(a = 150, b = 180, c = 210)
    PL: Open right (a = 180, b = 210) 
    
Rules FOR cruise
R1: if SD is NL and A is ZE then TC is PL
R2: if SD is ZE and A is NL then TC is PL
R3: if SD is NM and A is ZE then TC is PM
R4: if SD is NS and A is PS then TC is PS
R5: if SD is PS and A is NS then TC is NS
R6: if SD is PL and A is ZE then TC is NL
R7: if SD is ZE and A is NS then TC is PS
R8: if SD is ZE and A is NM then TC is PM
R9: if SD is ZE and A is ZE then TC is ZE

Rules FOR Lane correction
R1: if D1 is NL and D2 is ZE then STEERING is PL
R2: if D1 is ZE and D2 is NL then STEERING is PL
R3: if D1 is NM and D2 is ZE then STEERING is PM
R4: if D1 is NS and D2 is PS then STEERING is PS
R5: if D1 is PS and D2 is NS then STEERING is NS
R6: if D1 is PL and D2 is ZE then STEERING is NL
R7: if D1 is ZE and D2 is NS then STEERING is PS
R8: if D1 is ZE and D2 is NM then STEERING is PM
R9: if D1 is ZE and D2 is ZE then STEERING is ZE

Rules FOR Anticollision safe system
R1: if D is NL and S is ZE then A is PL
R2: if D is ZE and S is NL then A is PL
R3: if D is NM and S is ZE then A is PM
R4: if D is NS and S is PS then A is PS
R5: if D is PS and S is NS then A is NS
R6: if D is PL and S is ZE then A is NL
R7: if D is ZE and S is NS then A is PS
R8: if D is ZE and S is NM then A is PM
R9: if D is ZE and S is ZE then A is ZE



"""
# Lane assist can also be done with the use of same fuzzy rules and inference here instead of acceleration and speed we will take distance with respect to two dotted lines in cm
import numpy as np
Speed = 100
Acceleration = 50
DistanceL1 = 60
DistanceL2 = 100
DISTANCE = 100
VELOCITY = 50
print("The distance from first line ", DistanceL1 )
print("The distance from second line ",DistanceL2 )
print("The speed input is: ", Speed)
print("The Acceleration input is: ", Acceleration)
print("The Distance of the vehicle in the front is: ", DISTANCE)
print("The Velocity of the vehicle in the front is: ", VELOCITY)
print("\n")

#Functions for open left-Right fuzzyfication  
def openLeft(x,alpha, beta):
    if x<alpha:
        return 1
    if alpha<x and x<=beta:
        return (beta - x)/(beta - alpha)
    else:
        return 0
    
def openRight(x,alpha, beta):
    if x<alpha:
        return 0
    if alpha<x and x<=beta:
        return (x-alpha)/(beta - alpha)
    else:
        return 0

# Function for traingular fuzzyfication  
def triangular(x,a,b,c):
    return max(min((x-a)/(b-a), (c-x)/(c-b)),0)

#Fuzzy Partition 
def partition(x):
    NL = 0;  NM = 0; NS = 0; ZE = 0; PS = 0; PM = 0; PL = 0
    
    if x> 0 and x<60:
        NL = openLeft(x,30,60)
    if x> 30 and x<90:
        NM = triangular(x,30,60,90)
    if x> 60 and x<120:
        NS = triangular(x,60,90,120)
    if x> 90 and x<150:
        ZE = triangular(x,90,120,150)
    if x> 120 and x<180:
        PS = triangular(x,120,150,180)
    if x> 150 and x<210:
        PM = triangular(x,120,150,180)
    if x> 180 and x<240:
        PL = openRight(x,180,210)
 
    return NL,NM,NS,ZE,PS,PM,PL;

# Getting fuzzy values for all the inputs for all the fuzzy sets
NLSD,NMSD,NSSD,ZESD,PSSD,PMSD,PLSD = partition(Speed)
NLAC,NMAC,NSAC,ZEAC,PSAC,PMAC,PLAC = partition(Acceleration)
NLD1,NMD1,NSD1,ZED1,PSD1,PMD1,PLD1 = partition(DistanceL1)
NLD2,NMD2,NSD2,ZED2,PSD2,PMD2,PLD2 = partition(DistanceL2)
NLD,NMD,NSD,ZED,PSD,PMD,PLD = partition(DISTANCE)
NLV,NMV,NSV,ZEV,PSV,PMV,PLV = partition(VELOCITY)

# Display the fuzzy values for all fuzzy sets
outPut = [[NLSD,NMSD,NSSD,ZESD,PSSD,PMSD,PLSD],
          [NLAC,NMAC,NSAC,ZEAC,PSAC,PMAC,PLAC]]
outPut1 = [[NLD1,NMD1,NSD1,ZED1,PSD1,PMD1,PLD1],[NLD2,NMD2,NSD2,ZED2,PSD2,PMD2,PLD2]]
outPut2 = [[NLV,NMV,NSV,ZEV,PSV,PMV,PLV],[NLD,NMD,NSD,ZED,PSD,PMD,PLD ]]

print("The fuzzy values of the crisp inputs1")
print(["NL","NM","NS","ZE","PS","PM","PLSD"])
print(np.round(outPut,2))
print("\n")
print("The fuzzy values of the crisp inputs2")
print(["NL","NM","NS","ZE","PS","PM","PLSD"])
print(np.round(outPut1,2))
print("\n")
print("The fuzzy values of the crisp inputs3")
print(["NL","NM","NS","ZE","PS","PM","PLSD"])
print(np.round(outPut2,2))
print("\n")

# Rules implementation
def compare(TC1, TC2):
    TC = 0
    if TC1>TC2 and TC1 !=0 and TC2 !=0:
        TC = TC2
    else:
        TC = TC1
    
    if TC1 == 0 and TC2 !=0:
        TC = TC2
        
    if TC2 == 0 and TC1 !=0:
        TC = TC1
        
    return TC


def rule(NLSD,NMSD,NSSD,ZESD,PSSD,PMSD,PLSD,NLAC,NMAC,NSAC,ZEAC,PSAC,PMAC,PLAC):
    PLTC1 = min(NLSD,ZEAC) 
    PLTC2 = min(ZESD,NLAC)
    PLTC = compare(PLTC1, PLTC2)
    
    PMTC1 = min(NMSD,ZEAC)
    PMTC2 = min(ZESD,NMAC)
    PMTC = compare(PMTC1, PMTC2)
    
    PSTC1 = min(NSSD,PSAC)
    PSTC2 = min(ZESD,NSAC)
    PSTC = compare(PSTC1, PSTC2)
    NSTC = min(PSSD,NSAC)
    NLTC = min(PLSD,ZEAC)
    
    return PLTC, PMTC, PSTC, NSTC, NLTC;

PLTC, PMTC, PSTC, NSTC, NLTC = rule(NLSD,NMSD,NSSD,ZESD,PSSD,PMSD,PLSD,NLAC,NMAC,NSAC,ZEAC,PSAC,PMAC,PLAC)
PLTC1, PMTC1, PSTC1, NSTC1, NLTC1 = rule(NLD1,NMD1,NSD1,ZED1,PSD1,PMD1,PLD1,NLD2,NMD2,NSD2,ZED2,PSD2,PMD2,PLD2)
PLTC2, PMTC2, PSTC2, NSTC2, NLTC2 = rule(NLV,NMV,NSV,ZEV,PSV,PMV,PLV,NLD,NMD,NSD,ZED,PSD,PMD,PLD)
print("\n")
# Display the fuzzy values for all rules
outPutRules = [[PLTC, PMTC, PSTC, NSTC, NLTC ]]
outPutRules1 = [[PLTC1, PMTC1, PSTC1, NSTC1, NLTC1 ]]
outPutRules2 = [[PLTC2, PMTC2, PSTC2, NSTC2, NLTC2 ]]
print("The fuzzy output1: ")
print(["PLTC", "PMTC", "PSTC", "NSTC", "NLTC"])
print(np.round(outPutRules,2))
print("\n")
print("The fuzzy output2: ")
print(["PLTC", "PMTC", "PSTC", "NSTC", "NLTC"])
print(np.round(outPutRules1,2))
print("\n")
print("The fuzzy output3: ")
print(["PLTC", "PMTC", "PSTC", "NSTC", "NLTC"])
print(np.round(outPutRules2,2))
print("\n")

# De-fuzzyfication
def areaTR(mu, a,b,c):
    x1 = mu*(b-a) + a
    x2 = c - mu*(c-b)
    d1 = (c-a); d2 = x2-x1
    a = (1/2)*mu*(d1 + d2)
    return a # Returning area

def areaOL(mu, alpha, beta):
    xOL = beta -mu*(beta - alpha)
    return 1/2*mu*(beta+ xOL), beta/2

def areaOR(mu, alpha, beta):
    xOR = (beta - alpha)*mu + alpha
    aOR = (1/2)*mu*((240 - alpha) + (240 -xOR))
    return aOR, (240 - alpha)/2 + alpha

def defuzzyfication(PLTC, PMTC, PSTC, NSTC, NLTC):
    areaPL = 0; areaPM = 0; areaPS = 0; areaNS = 0; areaNL = 0;
    cPL = 0; cPM = 0; cPS = 0; cNS = 0; cNL = 0;

    if PLTC != 0:
        #areaPL, cPL = areaOR(PLTC, 180, 210)
        areaPL, cPL = areaOR(PLTC, 180, 210)
                
    if PMTC != 0:
        areaPM = areaTR(PMTC, 150, 180, 210)
        cPM = 180
    
    if PSTC != 0:
        areaPS = areaTR(PSTC, 120, 150, 180)
        cPS = 150
          
    if NSTC != 0:
        areaNS = areaTR(NSTC, 60, 90, 120)
        cNS = 90
        
    if NLTC !=0:
        areaNL, cNL = areaOL(NLTC, 30, 60)
        
    numerator = areaPL*cPL + areaPM*cPM + areaPS*cPS + areaNS*cNS + areaNL*cNL
    denominator = areaPL + areaPM + areaPS + areaNS + areaNL
    if denominator ==0:
        print("No rules exist to give the result")
        return 0
    else:
        crispOutput = numerator/denominator
        return crispOutput
crispOutputFinal = defuzzyfication(PLTC, PMTC, PSTC, NSTC, NLTC)
crispOutputFinal1 = defuzzyfication(PLTC1, PMTC1, PSTC1, NSTC1, NLTC1)
crispOutputFinal2 = defuzzyfication(PLTC2, PMTC2, PSTC2, NSTC2, NLTC2)

if crispOutputFinal !=0:
    print("\nThe crisp TC value is: ", crispOutputFinal)
    print("\nThe crisp STEERING value is: ", crispOutputFinal1)
    print("\nThe crisp ACCELERATION value is: ", crispOutputFinal2)

  
# Lane controller
distance1 = 11  # distance with respect to the left of the vehicle
distance2 = 2   # distance with respect to right of the vehicle 
speed = 67
def lane(distance1,distance2,speed):
     if(distance2<5):
      print("Continue in the same lane");
     else:  
      if(distance1<10 and speed<50) :
       print("continue in your lane ")
      elif(distance1<10 and speed<100):
       print("Move towards middle lane")
      elif(distance1<10):
       print("Move towards middle lane")
      elif(distance1<20 and speed<50):
       print("Move towards left lane")
      elif(distance1<20 and speed<100):
       print("Continue in your lane")
      elif(distance1<20 ):
       print("Move towards right lane")
      elif(speed<50):
       print("Move towards left lane")
      elif(speed<100):
       print("Move towards middle lane")
      elif(speed>100):
       print("Continue in your lane")

lane(distance1,distance2,speed)




```
